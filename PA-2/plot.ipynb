```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("checkpoints/HED_3.csv")
data
```

```python
plt.figure(figsize=(5,5))
plt.plot(data["Epoch"], data["Train Loss"], 'r.-' , label="Training Loss")
plt.plot(data["Epoch"], data["Validation Loss"], 'g.-' , label="Validation Loss")
plt.xlim(0, int(data["Epoch"][-1:-2:-1]))
plt.title("Learning Curves", fontsize=16)
plt.legend()
plt.xlabel("Epochs")
plt.ylabel("Loss per epoch")
plt.grid()
```
